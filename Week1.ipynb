**OOP's Basics**

```python
class Employee:

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary

    def display(self):
        print(self.name,self.salary)

e1 = Employee("Ramya",50000)
e1.display()
```

**SQL using Python**

```python
import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()
```

Create Table

```python
cursor.execute("""
CREATE TABLE employees(
id INTEGER PRIMARY KEY,
name TEXT,
age INTEGER,
salary INTEGER
)
""")
```

Insert Data


```python
cursor.execute("INSERT INTO employees VALUES(1,'Ramya',22,50000)")
cursor.execute("INSERT INTO employees VALUES(2,'Keerthi',23,60000)")
conn.commit()
```

Select Data

```python
cursor.execute("SELECT * FROM employees")
print(cursor.fetchall())
```

Where condition

```python
cursor.execute("SELECT * FROM employees WHERE salary > 55000")
print(cursor.fetchall())
```

Update Data

```python
cursor.execute("UPDATE employees SET salary = 70000 WHERE id = 1")
conn.commit()
```

**Python + SQL Integration**

```python
def add_employee(id,name,age,salary):
    cursor.execute("INSERT INTO employees VALUES(?,?,?,?)",(id,name,age,salary))
    conn.commit()

add_employee(4,"Sri",24,55000)
```
